```python
import numpy as np
import matplotlib.pyplot as plt

R_x = np.array([[1, 0.8182], [0.8182, 1]])
r_xd = np.array([0.8182, 0.354])

def gradient_descent(R_x, r_xd, lr, num_iterations):
    # Initialize w at zero
    w = np.zeros(2)
    log = [w.copy()]
    
    for n in range(num_iterations):
        gradient = np.dot(R_x, w) - r_xd  # Compute gradient
        w = w - lr * gradient  # Update weights
        log.append(w.copy())  # Store the updated weights
    
    return np.array(log)

# Set learning rates and number of iterations
lr_1 = 0.3
lr_2 = 1
num_iterations = 100

# Perform gradient descent for both learning rates
trajectory_eta_1 = gradient_descent(R_x, r_xd, lr_1, num_iterations)
trajectory_eta_2 = gradient_descent(R_x, r_xd, lr_2, num_iterations)

# Plotting function
def plot_trajectory(trajectory, eta, title):
    plt.plot(trajectory[:, 0], trajectory[:, 1], marker='o', label=f'Learning Rate: {eta}')
    plt.title(title)
    plt.xlabel('w1')
    plt.ylabel('w2')
    plt.grid(True)

# Plot the trajectories for both learning rates
plt.figure(figsize=(12, 6))

plot_trajectory(trajectory_eta_1, lr_1, 'Gradient Descent Trajectory (Learning rate = 0.3)')

plot_trajectory(trajectory_eta_2, lr_2, 'Gradient Descent Trajectory (Learning rate = 1)')

plt.legend()
plt.show()
```
