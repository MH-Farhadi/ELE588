```python
import numpy as np
import matplotlib.pyplot as plt


```

```python
def gradient_descent(R_x, r_xd, lr, num_iterations):
    w = np.zeros(2)
    log = [w.copy()]
    
    for n in range(num_iterations):
        gradient = np.dot(R_x, w) - r_xd 
        w = w - lr * gradient  
        log.append(w.copy())  
    
    return np.array(log)
```

```python
def plot_weights_over_epochs(trajectory, eta, title):
    epochs = np.arange(len(trajectory))
    plt.plot(epochs, trajectory[:, 0], label='w1')
    plt.plot(epochs, trajectory[:, 1], label='w2')
    plt.title(title)
    plt.xlabel('Epochs (Iterations)')
    plt.ylabel('Weight Values')
    plt.grid(True)
    plt.legend()
```

```python
R_x = np.array([[1, 0.8182], [0.8182, 1]])
r_xd = np.array([0.8182, 0.354])
```

```python

lr_1 = 0.3
lr_2 = 1
num_iterations = 50
trajectory_eta_1 = gradient_descent(R_x, r_xd, lr_1, num_iterations)
trajectory_eta_2 = gradient_descent(R_x, r_xd, lr_2, num_iterations)
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plot_weights_over_epochs(trajectory_eta_1, lr_1, 'Learning rate = 0.3')
plt.subplot(1, 2, 2)
plot_weights_over_epochs(trajectory_eta_2, lr_2, 'Learning rate = 1')
plt.tight_layout()
plt.savefig("weights.png", dpi=300)
plt.show()
```
